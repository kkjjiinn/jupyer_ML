# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist] # 폰트 이름만 가져오기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지지 않게
x=[1,2,3]
y=[2,4,8]
plt.plot(x,y)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth=10) #라인 두께 설정
```

```python
plt.plot(x,y, marker='o')#그래프에 마커 찍기
```

```python
plt.plot(x,y, marker='o', linestyle='None')#라인 없애기
```

```python
plt.plot(x,y, marker='v', markersize=10)#마커 v모양 # matplotlib marker검색
```

```python
plt.plot(x,y, marker='v', markersize=10, markeredgecolor='red')#마커 색깔 바꾸기
```

```python
plt.plot(x,y, marker='v', markersize=10, markeredgecolor='red', markerfacecolor='yellow')
```

## 선스타일


```python
plt.plot(x,y, linestyle=':')
```

```python
plt.plot(x,y, linestyle='--')
```

```python
plt.plot(x,y, linestyle='-.') # matplotlib linestyle
```

## 색깔

```python
plt.plot(x,y, color='b') #matbplotlib color
```

## 포맷

```python
plt.plot(x,y,'ro--')# color, marker, linestyle
```

```python
plt.plot(x,y, 'bv:')
```

```python
plt.plot(x,y, 'bv') # = plt.plot(x,y,color='b', marker='v', linestyle='None')
```

## 축약어

```python
plt.plot(x,y,marker='o', mfc='red',ms=10, mec='blue', ls=':')
```

## 투명도

```python
plt.plot(x,y,marker='o', mfc='red',ms=10, alpha=0.1)# alpha : 투명도 (0~1)
```

## 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10), dpi=10)# dots per inch확대
plt.plot(x,y)
```

# 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x,y)
```
