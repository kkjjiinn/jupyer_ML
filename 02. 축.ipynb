# 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist] # 폰트 이름만 가져오기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지지 않게
```

```python
x=[1,2,3]
y=[2,4,8]
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('x축', color='red')
plt.ylabel('y축',color='#00aa00')
```

```python
plt.plot(x,y)
plt.xlabel('x축', color='red',loc='right')# left, center, right
plt.ylabel('y축',color='#00aa00',loc='top')# top, center, botton
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.show()
```
