# 2. 출생아 수 및 합계 출산율

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist] # 폰트 이름만 가져오기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지지 않게
```

```python
import pandas as pd
df= pd.read_excel('합계출산율.xlsx', skiprows = 2,nrows=2,index_col=0)
df.iloc[1] = df.iloc[1].str.replace(',','').astype(int)
df.iloc[1] // 1000
```

```python
df=df.T
df
```

```python
plt.plot(df.index, df['출생아 수'])
plt.plot(df.index, df['합계출산율'])
```

```python
fig, ax1 = plt.subplots(figsize=(10,7))
ax1.plot(df.index, df['출생아 수'] // 1000,color='y')

ax2 = ax1.twinx() #x축을 공유하는 쌍둥이 axis
ax2.plot(df.index, df['합계출산율'],color='b')
```

```python
fig, ax1 = plt.subplots(figsize=(10,7))
#ax1.set_ylim(300,500)#y축 범위 제한
#ax1.set_yticks([300,400,500])#y축 숫자 표시 제한
ax1.plot(df.index, df['출생아 수'],color='y')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val + 12, val, ha='center') # 그래프 위 밸류 표시

ax2 = ax1.twinx() #x축을 공유하는 쌍둥이 axis
#ax2.set_ylim(0,1.5)
#ax2.set_yticks([0,1])
ax2.plot(df.index, df['합계출산율'],color='b', marker='o', ms=15, lw=5, mec='w', mew=3)
```
